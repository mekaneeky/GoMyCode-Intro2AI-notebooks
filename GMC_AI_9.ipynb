```
import math
import mlxtend
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
 

```

```
dataset = [['Skirt', 'Sneakers', 'Scarf', 'Pants', 'Hat'],
        ['Sunglasses', 'Skirt', 'Sneakers', 'Pants', 'Hat'],
        ['Dress', 'Sandals', 'Scarf', 'Pants', 'Heels'],
        ['Dress', 'Necklace', 'Earrings', 'Scarf', 'Hat', 'Heels', 'Hat'],
      ['Earrings', 'Skirt', 'Skirt', 'Scarf', 'Shirt', 'Pants']]

```

```
te=TransactionEncoder()
te_ary=te.fit(dataset).transform(dataset)    #Apply one-hot-encoding on our dataset
tx_df=pd.DataFrame(te_ary, columns=te.columns_)  #Creating a new DataFrame from our Numpy array
```

```
frequent_itemsets = apriori(tx_df, min_support=0.5, use_colnames=True)
```

```
association_rules(frequent_itemsets,metric="lift",min_threshold=1.00)
```

# Proposed Strategy 1

It appears that "lift" is greater than 1.0 indicating that the Skirts and Pants are more likely to sell together. I propose that a discount would be made on the bundle Skirt + Pants to increase sales.



```
association_rules(frequent_itemsets,metric="confidence",min_threshold=0.7) # associate itemsets with confidence over 70%.
```

```
big_tx_df = pd.read_csv("Market_Basket_Optimisation.csv", header=None)
```

```
big_tx_df.isnull().sum()
```

```
big_tx_lol = big_tx_df.values.tolist()

### Clean NaNs
cleaned_tx_lol = []
for tx_list in big_tx_lol:
    cleaned_tx_list = []
    for tx in tx_list:
        if type(tx) != str:
            continue
        else:
            cleaned_tx_list.append(tx)
    cleaned_tx_lol.append(cleaned_tx_list)

te=TransactionEncoder()
te_ary=te.fit(cleaned_tx_lol).transform(cleaned_tx_lol)    #Apply one-hot-encoding on our dataset
cleaned_tx_df=pd.DataFrame(te_ary, columns=te.columns_)
```

```
cleaned_tx_df.head()
```

```
frequent_itemsets = apriori(cleaned_tx_df, min_support=0.03, use_colnames=True)
```

```
association_rules(frequent_itemsets,metric="lift",min_threshold=1.00)
```

```
association_rules(frequent_itemsets,metric="confidence",min_threshold=0.3) # associate itemsets with confidence over 70%.
```
